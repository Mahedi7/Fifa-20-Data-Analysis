```python
#loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
#finding out all columns name
for col in fifa.columns:
    print(col)
```

```python
fifa.shape
```

```python
#No. of cells
18278 * 104
```

```python
#finding out which country has the most No. of players
fifa['nationality'].value_counts()
```

```python
#finding out top 5 country with most No. of players
fifa['nationality'].value_counts()[0:5]
```

```python
#finding out only the names of the top 5 country with most No. of players
fifa['nationality'].value_counts()[0:5].keys()
```

```python
#making a bar plot
plt.figure(figsize = (5, 5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), color = 'g')
plt.xlabel('Countries')
plt.ylabel('No. of Players')
plt.show()
```

```python
#players name and salary
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
player_salary
```

```python
player_salary.head()
```

```python
#sorting players salary in descending order
player_salary = player_salary.sort_values(by = ['wage_eur'], ascending = False)
```

```python
player_salary
```

```python
player_salary.head()
```

```python
#making a bar plot
plt.figure(figsize = (8, 5))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color = ['skyblue', 'darkred', 'red', 'darkblue', 'darkred'])
plt.xlabel('Players Name')
plt.ylabel('Playes Salary')
plt.show()
```
